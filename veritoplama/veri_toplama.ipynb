```python
# Çekilen verilerin kaydedileceği gerekli listelerin oluşturulması
urlList = []
priceList = []
featTitles = []
featValues = []
```

```python
#for döngüsü ile sayfalardaki verilerin ve özelliklerin çekilmesi
for i in range(1,10):
    print(i)
    url = "https://www.otokocikinciel.com/ikinci-el-araba?sayfa=" + str(i) 
    req = requests.get(url, headers={
"User-Agent" : "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36"
})
    soup = BeautifulSoup(req.content, 'lxml')
    
    #beautifulsoup ile araç ilan linklerinin bulunması
    urls = soup.find_all("div",attrs={"class":"ok-spcrl-carListBox"})
    for url in urls:
        car_url = "https://www.otokocikinciel.com"+url.a.get("href")
        urlList.append(car_url) #linklerin kaydedilmesi
        print(car_url)
        
        req_car = requests.get(car_url, headers={ #bulunan linklere istek atılması
"User-Agent" : "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36"
})
        soup_car = BeautifulSoup(req_car.content, 'lxml') #veri içeriğinin çekilmesi
        
        #araç özelliklerinin bulunduğu tablodan çekilmesi
        features = soup_car.find_all("table")
        table = soup_car.find('table')
        for row in table.find_all('tr'):
    
            columns = row.find_all('td')
    
            featTitles.append(columns[0].text)
            featValues.append(columns[1].text)
            
    #arac fiyatlarının bulunması ve listeye kaydedilmesi    
    prices = soup.find_all("div", attrs={"class":"ok-veh-price"})
    for price in prices:
        priceList.append(price.text.strip())
        print(price.text.strip())

#bulunan veri adetlerinin yazdırılması
print(str(len(urlList))+" adet link bulundu")
print(str(len(priceList))+" adet fiyat bulundu")
print(str(len(featTitles))+" adet özellik başlığı bulundu")
print(str(len(featValues))+" adet özellik verisi bulundu")
    
```

```python
#bulunan özelliklerin yazdırılması
for ozellik, deger in zip(featTitles, featValues):
    print(f"Özellik: {ozellik}, Değer: {deger}")
```

```python
#Url ve fiyatların dataframe'e yazılması
df_urls = pd.DataFrame()
df_urls["urls"] = urlList
df_urls["prices"] = priceList
```
