```python
#gerekli kütüphanelerin import edilmesi
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#pd.set_option('display.max_row', 705)
#pd.set_option('display.max_column', 15)
```

```python
#csv dosyasından verilerin okunması
data = pd.read_csv("car_data.csv")
```

```python
data
```

```python
#Kullanılacak özelliklerin başka bir dataframe'e isimleri ingilize karakterlere uyularak alınması
df = pd.DataFrame()
df['Kilometre'] = data['Kilometre']
df['Marka'] = data['Marka']
df['Model'] = data['Model']
df['Model_Yili'] = data['Model Yılı']
df['Motor_Hacmi'] = data['Motor Hacmi']
df['Renk'] = data['Renk']
df['Vites_Tipi'] = data['Vites Tipi']
df['Yakit_Turu'] = data['Yakıt Türü']
df['Cekis_Tipi'] = data['Çekiş Tipi']
df['Fiyat'] = data['price']
```

```python
#dataframe bilgi
df.info()
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.shape
```

```python
#veriseti ile ilgili bilgiler
df.info()
```

```python
df.isna().sum()
```

```python
df["Fiyat"] = df.Fiyat.str.replace('TL','')
df["Fiyat"] = df.Fiyat.str.replace('.','')
df["Motor_Hacmi"] = df.Motor_Hacmi.str.replace(',','')
```

```python
df
```

```python
df["Fiyat"] = df.Fiyat.astype(float)
df["Motor_Hacmi"] = df.Motor_Hacmi.astype(float)
```

```python
df
```

```python
df["Marka"] = df.Marka.str.replace(' ','_')
list_marka = df["Marka"].unique()
list_marka.sort()
```

```python
df_markalar = pd.DataFrame()
df_markalar["Markalar"] = list_marka
df_markalar
```
