```python
# veri analiz kütüphaneleri 
import pandas as pd
import numpy as np

# Görselleştirme
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# uyarıları ihmal etmek  (aktif değil incelersiniz)
import warnings
warnings.filterwarnings('ignore')
```

```python
# Veriyi import etme data
df = pd.read_csv('car_data_processed.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
#verinin matplotlib kütüphanesi ile görselleitirilmesi
columns = ['Kilometre', 'Marka', 'Model_Yili', 'Motor_Hacmi', 'Vites_Tipi',
           'Kilometre', 'Yakit_Turu', 'Cekis_Tipi', 'Fiyat']
#sütunlardaki verilerin tablo halinde yazdırılması
def distplot(nrows, ncols, columns):
    
    rows=nrows
    cols=ncols
    
    fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(18,12))
    
    columns = columns
    index=0
    
    for i in range(rows):
        for j in range(cols):
            sns.distplot(df[columns[index]], ax=ax[i][j], bins=40)
            index+=1
            
distplot(3, 3, columns)
```

```python
#Fiyat verilerinin tablosu
plt.figure()
plt.hist(df['Fiyat'], bins=150)
plt.show()
```

```python
# Özniteliklerin korelasyonunu heatmap ile gösterilmesi
plt.figure(figsize=(15,7))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu")
```
